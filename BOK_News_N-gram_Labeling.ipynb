```
import pandas as pd
from google.colab import drive
import datetime
```

```
data_1 = pd.read_csv('/content/drive/My Drive/자연어 처리 1조/데이터셋/news_handle.csv', encoding = 'utf-8')
```

```
data_1
```

```
data_1.drop(['Unnamed: 0'], axis='columns', inplace=True)
```

```
data_1
```

```
data_1['new_date'] = data_1.new_date.apply(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d"))
data_1
```

```
ir = pd.read_csv('/content/drive/My Drive/자연어 처리 1조/데이터셋/interest_rate.csv', encoding='utf-8')
ir.rename(columns = {'Unnamed: 0' : 'date'}, inplace = True)
ir['new_date'] = ir.date.apply(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d"))
ir.drop(['date', '기준 금리'], axis=1, inplace=True)
ir['before_month'] = ir['new_date']
ir.head()
```

```
ir
```

```
# 1. new_date에 대해 콜금리 컬럼 생성
data = pd.merge(data_1, ir, on='new_date')

# 2. before_month에 대해 콜금리 컬럼 생성
data.before_month = data.new_date.apply(lambda x: x - datetime.timedelta(days=30))
data = pd.merge(data, ir, on='before_month')

# 3. 콜금리 변화 컬럼(change) 생성
data['change'] = data['콜 금리_x'] - data['콜 금리_y']
data
```

```
# 콜금리 변화에 따른 라벨링

def labeling(change):
    if change > 0.03:
        return 1
    elif change < -0.03:
        return -1
    else:
        return 0
        
data['label'] = data.change.apply(lambda x: labeling(x))
data
```

```
data.drop(['new_date_x', '콜 금리_x', 'before_month', '콜 금리_y', 'new_date_y', 'change'], axis='columns', inplace=True)
data
```

```
data.to_csv(r'/content/drive/My Drive/자연어 처리 1조/데이터셋/news_handle_ir.csv')
```
